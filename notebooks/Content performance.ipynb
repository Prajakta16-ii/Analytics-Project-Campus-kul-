```python
import pandas as pd
import numpy as np

users = pd.read_csv("USER.csv")   # your existing file
users.head()
```

```python
content_types = [
    'Posts',
    'News',
    'Notices',
    'Resources',
    'Events',
    'Jobs',
    'Achievements'
]
```

```python
num_records = 5000

content = pd.DataFrame({
    'content_id': range(1, num_records + 1),
    
    # Randomly assign user_id from existing users
    'user_id': np.random.choice(users['user_id'], num_records),
    
    # Random content type
    'content_type': np.random.choice(content_types, num_records),
    
    # Random engagement
    'views': np.random.randint(20, 500, num_records),
    'comments': np.random.randint(0, 50, num_records),
    
    # Random created date (last 30 days)
    'created_at': pd.to_datetime('2026-02-01') + 
                  pd.to_timedelta(np.random.randint(0, 30, num_records), unit='D')
})

# Updated_at same as created_at initially
content['updated_at'] = content['created_at']

content.head()
```
