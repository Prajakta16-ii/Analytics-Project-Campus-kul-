```python
import pandas as pd

df = pd.read_csv("events.csv")
df.head()
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.drop_duplicates(subset=['event_id'])
```

```python
df
```

```python
#Remove rows with missing important columns
df = df.dropna(subset=['event_id', 'user_id', 'created_at'])
```

```python
df
```

```python

date_cols = ['start_date', 'end_date', 'created_at', 'updated_at']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df[col]
```

```python
df = df.dropna(subset=['created_at'])
```

```python
df
```

```python
df = df[df['updated_at'] >= df['created_at']]
df = df[df['end_date'] >= df['start_date']]
```

```python
df
```

```python
df['event_type'] = df['event_type'].str.lower().str.strip()
df['event_name'] = df['event_name'].str.strip()
df['location'] = df['location'].str.strip()
```

```python
df['event_type']
```

```python
df
```

```python
df.info()
```

```python
#remove extra space and line bars
df['event_description'] = df['event_description'].str.strip()
```

```python

df['event_description']
```

```python
#handles missing values
df['event_description'] = df['event_description'].fillna('not provided')
```

```python
df['event_description']
```

```python
df.to_csv("events_cleaned.csv", index=False)
```
