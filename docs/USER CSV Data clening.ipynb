```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("USER.csv")   # or pd.read_excel("file.xlsx")
```

```python
df
```

```python
duplicate_count = df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df
```

```python
missing_values = df.isnull().sum()
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
```

```python
missing_percent
```

```python
df = df.fillna("Unknown")  # Example
# OR
df = df.dropna()
```

```python
df
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')
```

```python
df['signup_date']
```

```python
invalid_dates = df['signup_date'].isnull().sum()
```

```python
invalid_dates
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

outlier_counts = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_counts[col] = len(outliers)
```

```python
numeric_cols 
```

```python
df['name'] = df['name'].str.strip().str.title()
```

```python
df['name'] 
```

```python
invalid_name = df[df['name'].str.contains(r'[^a-zA-Z ]', regex=True)]
```

```python
invalid_name
```

```python
df['email'] = df['email'].str.strip().str.lower()
```

```python
df['email']
```

```python
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
invalid_email = df[~df['email'].str.match(email_pattern)]
```

```python
df['phone_number'] = df['phone_number'].astype(str)
df['phone_number'] = df['phone_number'].str.replace(r'\D', '', regex=True)

phone_pattern = r'^[6-9]\d{9}$'
invalid_phone = df[~df['phone_number'].str.match(phone_pattern)]
```

```python
df['college_name'] = df['college_name'].str.strip().str.title()
```

```python
df['college_name'].unique()
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'], format='%d-%m-%Y', errors='coerce')
df['updated_at'] = pd.to_datetime(df['updated_at'], format='%d-%m-%Y', errors='coerce')
```

```python
invalid_dates = df[df['updated_at'] < df['signup_date']]
```

```python
df['status'] = df['status'].str.strip().str.title()

df['status'] = df['status'].replace({
    'Inactiv': 'Inactive'
})
```

```python

df['user_id'].duplicated().sum()
```

```python
df.to_csv("cleaned_USER_data.csv", index=False)
```

```python
report = pd.DataFrame({
    "Column_Name": df.columns,
    "Data_Type": df.dtypes.values,
    
    "Missing_Count": df.isnull().sum().values,
    
    "Duplicate_Count": [df[col].duplicated().sum() for col in df.columns],
    "Unique_Values": df.nunique().values
})
```

```python
report["Invalid_Email_Count"] = report["Column_Name"].apply(
    lambda x: len(invalid_email) if x == "email" else 0
)

report["Invalid_Phone_Count"] = report["Column_Name"].apply(
    lambda x: len(invalid_phone) if x == "phone_number" else 0
)


```

```python
report.to_csv("data_quality_report.csv", index=False)
```
