```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("comments.csv")
```

```python
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
#Remove rows with empty comments
df = df.dropna(subset=["comment_text"])
```

```python
df
```

```python
#clean comment text

#1.convert to string
df["comment_text"] = df["comment_text"].astype(str)
```

```python
df["comment_text"]
```

```python

# Remove leading & trailing spaces
df["comment_text"] = df["comment_text"].str.strip()
```

```python
df["comment_text"] 
```

```python
# Remove extra spaces inside text
df["comment_text"] = df["comment_text"].str.replace(r"\s+", " ", regex=True)
```

```python
df["comment_text"] 
```

```python

# Convert to lowercase (optional but recommended)
df["comment_text"] = df["comment_text"].str.lower()
```

```python
df["comment_text"] 
```

```python
df = df[df["comment_text"] != ""]
```

```python
df
```

```python
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

# ===============================
# STEP 3: Find rows with 00:00:00
# ===============================
mask = (
    (df["created_at"].dt.hour == 0) &
    (df["created_at"].dt.minute == 0) &
    (df["created_at"].dt.second == 0)
)

# ===============================
# STEP 4: Generate random time
# ===============================
random_hours = np.random.randint(1, 24, size=mask.sum())
random_minutes = np.random.randint(0, 60, size=mask.sum())
random_seconds = np.random.randint(0, 60, size=mask.sum())

# ===============================
# STEP 5: Replace time ONLY
# ===============================
df.loc[mask, "created_at"] = (
    df.loc[mask, "created_at"]
    + pd.to_timedelta(random_hours, unit="h")
    + pd.to_timedelta(random_minutes, unit="m")
    + pd.to_timedelta(random_seconds, unit="s")
)

# ===============================
# STEP 6: Save cleaned file
# ===============================
df.to_csv("comments_time_fixed.csv", index=False)

print("✅ 00:00:00 replaced with random HH:MM:SS and saved")
```

```python
df["updated_at"] = pd.to_datetime(df["updated_at"], errors="coerce")

# ===============================
# STEP 3: Find rows with 00:00:00
# ===============================
mask = (
    (df["updated_at"].dt.hour == 0) &
    (df["updated_at"].dt.minute == 0) &
    (df["updated_at"].dt.second == 0)
)

# ===============================
# STEP 4: Generate random time
# ===============================
random_hours = np.random.randint(1, 24, size=mask.sum())
random_minutes = np.random.randint(0, 60, size=mask.sum())
random_seconds = np.random.randint(0, 60, size=mask.sum())

# ===============================
# STEP 5: Replace ONLY the time part
# ===============================
df.loc[mask, "updated_at"] = (
    df.loc[mask, "updated_at"]
    + pd.to_timedelta(random_hours, unit="h")
    + pd.to_timedelta(random_minutes, unit="m")
    + pd.to_timedelta(random_seconds, unit="s")
)

# ===============================
# STEP 6: Save cleaned file
# ===============================
df.to_csv("comments_updated_time_fixed.csv", index=False)

print("✅ updated_at time fixed and file saved")
```
